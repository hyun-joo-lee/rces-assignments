### 5) Extract the state or country using Pandas [text data functions](https://pandas.pydata.org/pandas-docs/stable/text.html)

Add it as a new column to the dataframe called `country`. Note that some of the "countries" are actually U.S. states.

```python
df['country'] = df['place'].str.split(", ").str[-1]
```

### 7) Make a bar chart of the top 5 earthquake magnitudes vs country/state 

```python
df_top50 = df.nlargest(5, ['mag'])
df_top50.plot(kind='bar', x = 'country', y = 'mag', xlabel = 'Country/State', ylabel = 'Magnitudes', title = 'Top 5 earthquakes by magnitude');
```

### 8) Create a filtered dataset that only has earthquakes of magnitude 4 or larger and

```python
df_large_4 = df[df.mag >= 4]
```

### 9) Make a histogram the distribution of the Earthquake magnitudes

https://pandas.pydata.org/pandas-docs/version/0.23/generated/pandas.DataFrame.hist.html
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html

Do one subplot for the filtered and one for the unfiltered dataset.
Use a Logarithmic scale. What sort of relationship do you see?

```python
fig, ax = plt.subplots(figsize = (12, 4), ncols = 2)
ax[0].hist(df['mag'], log = True)
ax[1].hist(df_large_4['mag'], log = True); 
```

Filtered data shows that the bigger earthquake occurs less, but unfiltered data shows the data of earthquake magnitude between 0 to 1 counted more than 2-4.

### 11) Visualize the locations of earthquakes by making a scatterplot of their latitude and longitude

Use a two-column subplot with both the filtered and unfiltered datasets. Color the points by magnitude. Make it pretty

What difference do you note between the filtered and unfiltered datasets?

```python
from matplotlib.colors import LogNorm

fig, ax = plt.subplots(figsize = (12, 4), ncols = 2)
ax[0].scatter(df['longitude'], df['latitude'], c = df['mag'], cmap = 'nipy_spectral', s = 7)
splot = ax[1].scatter(df_large_4['longitude'], df_large_4['latitude'], c = df_large_4['mag'], cmap = 'nipy_spectral', s = 7, vmin = df['mag'].min(), vmax = df['mag'].max())

ax[0].set_xlabel('Longitude (°)')
ax[0].set_ylabel('Latitude (°)')
ax[0].grid(which = 'major', linewidth = 0.5)
ax[0].set_title('Magnitude 0~')

ax[1].set_xlabel('Longitude (°)')
ax[1].set_ylabel('Latitude (°)')
ax[1].grid(which = 'major', linewidth = 0.5)
ax[1].set_title('Magnitude 4~')

fig.suptitle('Location of EQ by magnitude')
fig.colorbar(splot).set_label('Magnitude');
```

Unfiltered data shows that very weak earthquake data is mostly on the United States and Canada, which means the weak EQ data is mostly collected on North America. The data collection itself has bias.
